```python
import pandas as pd
import matplotlib.pyplot as plt

kpi_df = pd.read_csv("data/gold/fraud_kpis.csv")

metrics = [
    "total_fraud_cases",
    "suspicious_transactions",
    "declined_transactions"
]

values = [
    kpi_df["total_fraud_cases"][0],
    kpi_df["suspicious_transactions"][0],
    kpi_df["declined_transactions"][0]
]

plt.figure(figsize=(8,5))
plt.bar(metrics, values)

plt.title("Fraud Analytics KPI Overview")
plt.ylabel("Count")

plt.show()
```

```python
merchant_df = pd.read_csv("data/gold/merchant_fraud_analysis.csv")

top_merchants = merchant_df.sort_values(
    by="fraud_cases",
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

plt.bar(
    top_merchants["merchant"],
    top_merchants["fraud_cases"]
)

plt.xticks(rotation=90)

plt.title("Top Merchants by Fraud Cases")
plt.ylabel("Fraud Cases")

plt.show()
```

```python
country_df = pd.read_csv("data/gold/country_risk_analysis.csv")

plt.figure(figsize=(8,5))

plt.bar(
    country_df["country"],
    country_df["fraud_cases"]
)

plt.title("Fraud Cases by Country")
plt.ylabel("Fraud Cases")

plt.show()
```
